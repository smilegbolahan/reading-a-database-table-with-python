```python
from pymysql import connect

database = connect(
                    host = "localhost",
                    user = "root",
                    passwd = "",
                    database = "test"
)

tool = database.cursor()
query = "show tables"
tool.execute(query)

#to show tables in the database
for table in tool:
    print(table)
```

```python
#to show the rows in the dataframe

query= "select * from customer"
tool.execute(query)
```

```python
import pandas as pd

staff = pd.read_sql(query,database)
display(staff)
```
